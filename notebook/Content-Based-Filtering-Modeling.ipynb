####  Importing Libraries

```python
import pandas as pd
import joblib
```

####  Reding the Input Files

```python
df_article=pd.read_csv('articles.csv')
```

### Saving the article file 

```python
joblib.dump(df_article,'articles.df')
```

```python
df_article.info()
```

### Feature Enginering by creating the dummy variables for below

```python
article=df_article[['article_id','product_group_name','garment_group_name','index_group_name']]
article.head()
```

```python
meta_data=pd.get_dummies(article,columns=['product_group_name','garment_group_name','index_group_name']).astype(int)
meta_data.head()
```

### Saving the above meta_data file

```python
meta_data.to_csv('meta_data.csv')
joblib.dump(df_article,'articles.df')
```

```python
meta_data.head(1)
```

```python
articleid='816166002'
article = df_article.index[df_article['article_id'] == int(articleid)]
article
```

```python
y=meta_data.loc[article]
y
```

```python
joblib.dump(meta_data,'meta_data.df')
```

### K-Nearest Neighbors Model

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(n_neighbors=5)
knn.fit(meta_data)
n_recs=5
# Return results using .kneighbors attribute of knn model
index = knn.kneighbors(X=y, n_neighbors=n_recs+1, return_distance=False).flatten()
results = df_article.iloc[index].index.values[1:]
results
```

### Saving the model file

```python
joblib.dump(knn,'knn.mt')
```

```python
# Using returned results variable, index the original meta data frame to return appropriate information for each article
results_df = df_article.loc[results]
results_df.reset_index(inplace=True)
results_df
```
